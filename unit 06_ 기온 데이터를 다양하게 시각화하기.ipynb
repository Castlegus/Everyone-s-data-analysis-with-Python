```python
# 8월의 기온만 뽑아서
import csv
import matplotlib.pyplot as plt

f = open('C:/Users/정성현/모두의 데이터분석/data/seoul.csv')
data = csv.reader(f)
next(data)
August = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1]  != '':
        if month == '08':
            August.append(float(row[-1]))
        
plt.hist(August, bins=100, color='r')
plt.show()

# 역대 8월 최고 기온의 빈도 수를 살펴보면 32도 정도가 가장 많았고, 20도 이하이거나 거의 40도에 육박한 적도 있었다.
```

```python
# 월별 최고 기온 박스 플롯으로 나타내기
import csv
import matplotlib.pyplot as plt

f = open('C:/Users/정성현/모두의 데이터분석/data/seoul.csv')
data = csv.reader(f)
next(data)
month = [[],[],[],[],[],[],[],[],[],[],[],[]]
# 반복문을 이용하여 month=[]값에 []를 12번 append 해도 된다.

for row in data:
    if row[-1] !='':
        month[int(row[0].split('-')[1])-1].append(float(row[-1])) # 월 데이터를 month의 인덱스에 맞추기 위해 -1 처리. 
                                                                    # (ex. 1월데이터 -> month의 0번째 인덱스에)

plt.boxplot(month)
plt.show()

# 위 아래의 o는 이상치를 표현한 것으로, 다른 수치에 비해 너무 크거나 작은 값을 자동으로 나타낸 것
```

```python
# 8월 일별 기온 데이터를 박스 플롯으로 나타내기

import csv
import matplotlib.pyplot as plt

f = open('C:/Users/정성현/모두의 데이터분석/data/seoul.csv')
data = csv.reader(f)
next(data)
day = []
for i in range(31):
    day.append([])

for row in data:
    if row[-1] !='':
        if row[0].split('-')[1] == '08':
            day[int(row[0].split('-')[2])-1].append(float(row[-1]))

plt.style.use('ggplot') # 그래프 스타일
plt.figure(figsize=(10,5), dpi=100) # 그래프 크기
plt.boxplot(day, showfliers=False) # 아웃라이어(이상치) 값 생략

plt.show()
```
