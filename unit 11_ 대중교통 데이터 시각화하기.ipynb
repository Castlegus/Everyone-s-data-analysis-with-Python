```python
# 유임 승차 비율이 가장 높은 역 찾기

import csv
f = open('C:/Users/정성현/모두의 데이터분석/data/subwayfee.csv', encoding='utf-8')
data = csv.reader(f)
next(data)

mx = 0
rate = 0
mx_station = ''

# 헤더를 제외하고 4번~7번 열을 정수(int)로 변환
for row in data:
    for i in range(4,8):
        row[i] = int(row[i]) 

    if row[4]+row[6] > 100000: # 유무임 승차 합 100000명 이상만
        rate = row[4] / (row[4]+row[6])
        if rate > mx:
            mx = rate
            mx_station = row[3] + '역' + ' ' + row[1] # 역이름 + 노선이름
            
print(mx_station + '의 유임 승차 비율이', mx, '로 가장 높다')
```

```python
# 유무임 승하차 인원이 가장 많은 역은 어디일까

import csv
f = open('C:/Users/정성현/모두의 데이터분석/data/subwayfee.csv', encoding='utf-8')
data = csv.reader(f)
next(data)

mx = [0 ,0, 0, 0]
mx_station = ['','','','']
label = ['유임승차','유임하차','무임승차','무임하차']

for row in data:
    for i in range(4,8):
        row[i] = int(row[i])
        
        if row[i] > mx[i-4]:
            mx[i-4] = row[i]
            mx_station[i-4] = row[3] + '역 ' + row[1]
            
for i in range(4):
    print(label[i],':', mx_station[i], mx[i],'명')
```
