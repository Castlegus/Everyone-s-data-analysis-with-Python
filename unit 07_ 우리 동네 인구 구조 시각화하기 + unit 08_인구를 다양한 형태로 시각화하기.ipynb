```python
# 행안부 홈페이지에서 2020년 6월 인구 공공데이터를 csv파일로 내려받습니다. 

import csv
f = open('C:/Users/정성현/모두의 데이터분석/data/age.csv')
data = csv.reader(f)

result=[]
for row in data:
    if '신도림' in row[0]: # 행정구역 명에 신도림이 포함되어 있다면
        for i in row[3:]: # 0세부터 끝(100세 이상)까지 모든 연령에 대하여
            result.append(int(i)) # result에 값을 int형으로 바꿔 추가한다.

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.plot(result)
plt.show()
```

```python
# input 값을 받아서 원하는 지역 인구구조 출력하기

import csv
f = open('C:/Users/정성현/모두의 데이터분석/data/age.csv')
data = csv.reader(f)

result=[]
name=input('지역(동)을 입력해주세요:')
for row in data:
    if name in row[0]:
        for i in row[3:]:
            result.append(int(i))
            
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font',family='Malgun Gothic')
plt.title('2020년 6월 '+ name+'의 인구 분포')
plt.plot(result)
plt.show()

# 그러나 예를들어 아현동을 input 하면 북아현동과 아현동 모두 포함되기에 그래프에 오류가 생긴다는 한계가 있다.
```

```python
# 남녀 성별 인구 구조 그래프 그리기

import csv
f=open('C:/Users/정성현/모두의 데이터분석/data/gender.csv')
data = csv.reader(f)
m=[]
f=[]

name=(input('지역(동)이름을 입력하세요:'))
for row in data:
    if name in row[0]:
        for i in range(0,101):
            m.append(int(row[i+3])*-1)
            f.append(int(row[-i-1]))
f.reverse() # 리스트의 값을 역순으로 재배열 하는 함수

import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.figure(figsize=(10,5), dpi=300)
plt.title(name+'의 남녀 성비 분포')
plt.barh(range(101),m, color='skyblue',label='남성')
plt.barh(range(101),f, color='pink',label='여성')
plt.legend()
plt.ylabel('나이')
plt.xlabel('인구수')
plt.show()
```
