```python
import numpy as np
print(np.random.choice(6,10)) # 0~5의 숫자 10개를 추출
print(np.random.choice(10,6, replace = False)) # 0~9의 숫자 중 6개 비복원추출
print(np.random.choice(6,10, p=[0.1,0.1,0.5,0.1,0.1,0.1])) # 0~5의 각각 확률을 정한다. 당연히 p의 합은 1이여야 함
```

```python
import csv
import numpy as np

f = open('C:/Users/정성현/모두의 데이터분석/data/age.csv')
data = csv.reader(f)
next(data)
data = list(data)

name = input('인구 구조가 알고 싶은 지역(동)의 이름은?: ')

mn=1 # 최솟값을 저장할 변수
result_name='' # 최솟값을 갖는 지역
result=0 #최솟값을 갖는 지역의 연령대별 인구 비율

for row in data:
    if name in row[0]: # 입력받은 지역의 이름이 포함되어 있다면,
        home = np.array(row[3:],dtype = int)/int(row[2]) # 연령대별 인구수를 그 지역의 전체 인구수로 나눈다.
        
for row in data:
    away = np.array(row[3:], dtype = int)/int(row[2]) # 다른 지역의 인구 비율
    s = np.sum((home-away)**2) # 음수 값을 없애기 위해 제곱
    
    if s <mn and name not in row[0]:# input으로 입력 받은 값은 제외한다.
        mn = s
        result_name = row[0]
        result = away
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
plt.title(name+ ' 지역과 가장 비슷한 인구구조를 가진 지역')
plt.plot(home, label = name)
plt.plot(result, label = result_name)
plt.legend()
plt.show()
```
