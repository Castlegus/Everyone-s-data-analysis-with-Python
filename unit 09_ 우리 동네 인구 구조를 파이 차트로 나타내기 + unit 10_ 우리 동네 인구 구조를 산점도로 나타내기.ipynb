```python
# 파이 차트 만들기 plt.pie

import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
size = [2441, 1232 , 3451, 5333]
label = ['가형','나형','다형','라형']
color = ['darkmagenta','deeppink','hotpink','pink']
# plt.axis('equal')
plt.pie(size, labels=label, autopct = '%.1f%%', colors = color, explode = (0,0,0.1,0)) 
# autopct 각 항목의 비율을 자동으로 계산해서 표시 , explode 돌출 정도
plt.legend
plt.show()
```

```python
import csv
f = open('C:/Users/정성현/모두의 데이터분석/data/gender.csv')
data = csv.reader(f)
size = []
name = input('궁금하신 동네를 입력하세요: ')
for row in data:
    if name in row[0]:
        m = 0
        f = 0
        for i in range(101):
            m+= int(row[i+3])
            f+=int(row[i+106])
        break
size.append(m)
size.append(f)
print(size)
# 결과를 보니 여성이 더 많다.
```

```python
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')
color = ['crimson', 'darkcyan']
plt.pie(size, labels=['남','여'],autopct='%.f%%',colors = color, startangle = 90)
# starangle 속성은 12시 정각 위치에서 그래프가 시작되게 한다.
plt.title(name+'지역의 남녀 성별 비율')
plt.show()
```

```python
# 꺾은선 그래프로 표현하기 plt.plot

import csv
f = open('C:/Users/정성현/모두의 데이터분석/data/gender.csv')
data = csv.reader(f)
m=[]
f=[]
name = input('궁금하신 동네를 입력하세요: ')
for row in data:
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i]))
            f.append(int(row[i+103]))
        break
import matplotlib.pyplot as plt
plt.plot(m, label='Male')
plt.plot(f, label='Female')
plt.legend()
plt.show()
```

```python
# 막대그래프로 표현하기 plt.bar

import csv
f = open('C:/Users/정성현/모두의 데이터분석/data/gender.csv')
data = csv.reader(f)
result=[]
name = input('궁금하신 동네를 입력하세요: ')
for row in data:
    if name in row[0]:
        for i in range(3,104):
            result.append(int(row[i])-int(row[i+103])) # 남성 수에서 여성 수를 뺀다.
        break
```

```python
import matplotlib.pyplot as plt
plt.bar(range(101), result)
plt.show()

# 대부분의 연령에서 여성의 수가 많다는 것을 알 수 있다.
```

```python
# 산점도로 표현하기 plt.scatter

import csv
f = open('C:/Users/정성현/모두의 데이터분석/data/gender.csv')
data = csv.reader(f)
m =[]
f =[]
name = input('궁금한 동네를 입력해주세요: ')
for row in data:
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i]))
            f.append(int(row[i+103]))
        break

import matplotlib.pyplot as plt
plt.scatter(m,f, c=range(101), alpha=0.5, cmap='jet') # c는 점의 색상(총 101개의 점), 
                                                      #alpha는 투명도 0~1, cmap은 컬러바의 색상 종류 (jet은 무지개색)
plt.colorbar()
plt.plot(range(max(m)),range(max(m)),'g') # 남자의 인구수를 기준으로 한 y = x 추세선
plt.show()

# 추세선의 위쪽에 있는 점들로 보아 역시 여성이 남성보다 많다는 것을 알 수 있다.
```

```python
import csv
import math
f = open('C:/Users/정성현/모두의 데이터분석/data/gender.csv')
data = csv.reader(f)
m =[]
f =[]
size=[]
name = input('궁금한 동네를 입력해주세요: ')
for row in data:
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i]))
            f.append(int(row[i+103]))
            size.append(math.sqrt(int(row[i])+int(row[i+103]))) # 점의 크기 설정. math.sqrt로 크기를 조금 작게 만들어준다.
        break
        
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font',family='Malgun Gothic')
plt.figure(figsize=(10,5), dpi = 200)
plt.title(name + '지역의 성별 인구 그래프')
plt.scatter(m,f,s=size ,c=range(101), alpha=0.5, cmap='jet')
plt.colorbar()
plt.plot(range(max(m)),range(max(m)),'g')
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.show()
```
