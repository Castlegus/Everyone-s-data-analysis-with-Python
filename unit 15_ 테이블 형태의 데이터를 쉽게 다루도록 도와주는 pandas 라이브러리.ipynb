```python
# 궁금한 지역의 인구구조와 가장 비슷한 지역들 찾기

df = pd.read_csv('C:/Users/정성현/모두의 데이터분석/data/age.csv', encoding='cp949',index_col=0)
df.head()
```

```python
df=df.div(df['2020년06월_계_총인구수'],axis=0) # 전체 데이터를 총 인구수로 나눈다.
del df['2020년06월_계_총인구수'], df['2020년06월_계_연령구간인구수'] # 사용하지 않는 열 삭제

name = input('원하는 지역(동)의 이름을 입력해주세요: ')
a= df.index.str.contains(name)
df2=df[a]
df2
```

```python
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')
df2.T.plot()
plt.show()
```

```python
import numpy as np

x = df.sub(df2.iloc[0], axis=1) # 궁금한 지역의 인구비율을 뺀다.
y = np.power(x,2) # 뺀 값의 제곱들을
z = y.sum(axis=1) # 모두 더한다.

i = z.sort_values().index[:5] # 그 차가 가장 작은 지역 5곳을 찾는다.

df.loc[i].T.plot()
plt.title(name+' 지역과 유사한 인구분포를 가진 지역들')
plt.xlabel('나이')
plt.ylabel('인구 비율')
plt.show()
```
