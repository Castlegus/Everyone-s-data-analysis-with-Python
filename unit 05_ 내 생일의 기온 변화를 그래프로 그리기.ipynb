```python
import csv
f = open('C:/Users// /data/seoul.csv')
data = csv.reader(f)
next(data)

result = []

for row in data:
    if row[-1] != '':
        result.append(float(row[-1])) #      result .
        
print(result)
```

```python
#     
import matplotlib.pyplot as plt
import csv
f = open('C:/Users// /data/seoul.csv')
data = csv.reader(f)
next(data)

high = []
low = []

for row in data:
    if row[-1] != '' and row[-2] != '': #       
            if 1995 <= int(row[0].split('-')[0]): #  1995  
                if row[0].split('-')[1] == '06' and row[0].split('-')[2] == '15': #,    06, 15
                    high.append(float(row[-1]))
                    low.append(float(row[-2]))

plt.plot(high , label = '')
plt.plot(low , label = '') # label plt.legend()  
plt.xticks(range(0,25)) # ..     .

plt.rc('font', family = 'Malgun Gothic')

plt.title('    ')

#   (-)    plt.rcParams['axes.unicode_minus'] = False   .

plt.legend()
plt.show()
```
