```python
import csv
f = open('C:/Users/정성현/모두의 데이터분석/data/seoul.csv')
data = csv.reader(f)
next(data)

result = []

for row in data:
    if row[-1] != '':
        result.append(float(row[-1])) # 누락값이 아닌 최고 기온 값들을 result에 넣는다.
        
print(result)
```

```python
print(len(result))
```

```python
import matplotlib.pyplot as plt
plt.plot(result, 'r') # 약 4만개의 데이터를 꺾은선 그래프로 표시
plt.show()
```

```python
# 그래프 크기 조절하기
plt.plot(result, 'r')
plt.figure(figsize = (10,2)) # 가로 10인치, 세로 2인치로 설정
plt.show() # 그러나 주피터 노트북 내에서는 같은 크기처럼 보인다.
```

```python
plt.plot(result, 'r')
plt.rcParams['figure.figsize'] = [100, 2] # rcParams을 이용하여 차트 그림(figure)의 기본 설정을 지정할 수 있다.
plt.show()
```

```python
# split 함수를 사용해서 8월의 최고 기온 데이터 추출해서 그리기
import matplotlib.pyplot as plt
import csv
f = open('C:/Users/정성현/모두의 데이터분석/data/seoul.csv')
data = csv.reader(f)
next(data)

result = []

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] =='08': # 만약 1번째 인덱스의 값이 08 이라면 (인덱스는 0번째 부터 시작)
            result.append(float(row[-1]))
            
plt.plot(result, 'hotpink')
plt.show()

# 하지만 이 그래프에서도 뭔가 의미를 찾기는 어렵다.
```

```python
# 내 생일의 기온변화 그래프 그리기
import matplotlib.pyplot as plt
import csv
f = open('C:/Users/정성현/모두의 데이터분석/data/seoul.csv')
data = csv.reader(f)
next(data)

high = []
low = []

for row in data:
    if row[-1] != '' and row[-2] != '': # 최고 기온 값과 최저 기온 값이 존재한다면
            if 1995 <= int(row[0].split('-')[0]): # 연도가 1995보다 크거나 같고
                if row[0].split('-')[1] == '06' and row[0].split('-')[2] == '15': #월, 일 값이 각각 06, 15라면
                    high.append(float(row[-1]))
                    low.append(float(row[-2]))

plt.plot(high , label = '최고기온')
plt.plot(low , label = '최저기온') # label은 plt.legend()의 범례에 사용
plt.xticks(range(0,25)) # 음.. 그냥 만 값으로 보는걸로 하자.

plt.rc('font', family = 'Malgun Gothic')

plt.title('내 생일의 기온 변화 그래프')

# 만약 그래프의 마이너스(-) 값이 깨져서 보인다면 plt.rcParams['axes.unicode_minus'] = False 를 적어주면 된다.

plt.legend()
plt.show()
```
